# Notebook showing how to access data 

```python
import rasterio
import geopandas as gpd
import ipyleaflet
import numpy as np
from eo_data_pipeline.data_access.access_layer import DataAccessLayer

import matplotlib.pyplot as plt
```

```python
# Load data for a specific item and band
item_path = "../data/raw/S2A_33TUL_20230110_0_L2A_nir.tif"

# Create a DataAccessLayer instance
data_access_layer = DataAccessLayer(item_path)

# Load the data
data = data_access_layer.load_data()[0,1000:1500,1000:1500] # I am just plotting 500x500 pixels

```

## Visualize the data 

```python

# Plot the data using matplotlib
plt.imshow(data, cmap='gray', origin='upper')
plt.xlabel('Column')
plt.ylabel('Row')
plt.colorbar()
plt.title("NIR Band")
plt.show()
```
